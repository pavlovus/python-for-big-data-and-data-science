5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Figure:
    def get_volume(self):
        return 0

    def __add__(self, other):
        return self.get_volume() + other.get_volume()

f = Figure()
print(f + f)
```

Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Cylinder(Figure):
    def __init__(self, radius, height):
        super().__init__()
        self.radius = radius
        self.height = height

    def get_volume(self):
        return self.radius*self.radius*math.pi*self.height

class Cube(Figure):
    def __init__(self, side):
        super().__init__()
        self.side = side

    def get_volume(self):
        return self.side**3

cylinder = Cylinder(1, 5)
cube = Cube(5)
print(cylinder + cube)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class  EWallet:
    def __init__(self, balance, id):
        self.__balance = balance
        self.__id = id

    @property
    def balance(self):
        return self.__balance

    @property
    def id(self):
        return self.__id

    @balance.setter
    def balance(self, value):
        if value < 0:
            raise ValueError("Balance cannot be negative")
        self.__balance = value

    def deposit(self, value):
        if value <= 0:
            raise ValueError("Deposit amount must be positive value")
        self.balance += value

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdraw amount must be positive")
        if amount > self.balance:
            raise ValueError("No enough balance to withdraw")
        self.balance -= amount

wallet = EWallet(100, 12345678)
print(wallet.balance)
wallet.deposit(100)
print(wallet.balance)
wallet.withdraw(50)
print(wallet.balance)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class ShoppingCart:
    def __init__(self, items):
        self.items = items

    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, name):
        for item in self.items:
            if item.name == name:
                self.items.remove(item)

    def print_items(self):
        if not self.items:
            print("No items in cart")
        for item in self.items:
            print(item.name + " " + str(item.price))

shopping_cart = ShoppingCart()
shopping_cart.add_item(Item("Beer", 30))
shopping_cart.add_item(Item("Snickers", 40))
shopping_cart.print_items()
shopping_cart.remove_item("Beer")
shopping_cart.print_items()
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class DayIterator:
    def __init__(self):
        self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= len(self.days):
            raise StopIteration
        day = self.days[self.i]
        self.i += 1
        return day

di = DayIterator()
for i in range(7):
    print(next(di))
```
