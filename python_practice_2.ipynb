15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

Обидві структури є впорядкованими послідовностями даних, вони можуть містити різні типи, до них можна звертатися через індекси і т.д. Головна відмінність список - змінний(mutable), кортеж - незмінний(immutable). Тобто, після створення кортежу звіти нічого не можна забрати, покласти чи змінити, а в список - можна.В списках зазвичай зберігаються дані, які часто змінюються, додаються чи видаляються, а в кортежах - ті, які ні в якому випадку не повинні змінитись. Також, кортежі сожуть бути ключами в словниках, а списки - ні.

# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
set1 = {1, 2, 3}
set1.add(4)
set1.add("Pavlo")
print(set1)
```

Так, множина є змінним типом даних, адже ми можемо додавати та видаляти дані з неї

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
set2 = set(my_list)
print(set2)
```

Бо множина - це колекція УНІКАЛЬНИХ елементів, а отже, всі повтори елементів зникають при перетворені списку на множину.

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set3 = {1, 2, 3, 4, 5}
set4 = {4, 5, 6, 7, 8}
print(set3.union(set4))
print(set3.difference(set4))
print(set3.intersection(set4))
print(set3.symmetric_difference(set4))
```

```python
dict1 = {}
dict1[1] = "int"
dict1["Pavlo"] = "str"
dict1[0.5] = "float"
dict1[True] = "bool"

print(dict1)
```
